#plotting basic graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

## resize your graph
plt.figure(figsize=(5,3), dpi = 100)

## using shorthand notation
##fmt='[color][marker][line]'
plt.plot(x,y, 'b*--', label='2x')

#line 1

#keyword argument notation
#plt.plot(x,y, label ='2x',color = 'red', linewidth=2, marker ='+' , markersize= 10, markedgecolor = 'blue')

##line number two

# select interval we want to plot points at
x2=np.arange(0,4.5,0.5)
print(x2)

#plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label = 'x^2')

#plot the remainder of the graph as dot
plt.plot(x2[5:], x2[5:]**2, 'r--') 

# add a title (specify font parameters with fontdict)
plt.title('our first graph', fontdict={'fontname': 'comic sans ms', 'fontsize': 20})

# x and y labels
plt.xlabel('x axis')
plt.ylabel('y axis') 

#x and y tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

# add a legend
plt.legend()

# save figure (dpi 300 is good when saving, so tha graph will have high resolution) 
plt.savefig('my_graph.png', dpi=300)
plt.show()
```

# Bar Chart

```python
labels = ['A', 'B' ,'C']
values = [1,4,2]
bars = plt.bar(labels, values)

patterns = ['*', 'o', '/']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
    
# alternative way of setting the hatch on each bars
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('*')
plt.figure(figsize=(6,4))
plt.show()
plt.savefig('first_bar_chart.png', dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas= pd.read_csv('gas_prices.csv')
```

```python
# to show all the country in the table
gas
for country in gas:
 if country != 'year':
    print(country)
    
    #to plot a graph containing all the countries
for country in gas:
    if country != 'Year':
     plt.plot(gas.Year, gas[country], marker='*',label= country)
    
    plt.xlabel('Year',fontdict={'fontsize':14})
    plt.ylabel('US Dollar(s)', fontdict={'fontsize':14})
   # to set the graph year to three years interval and icluding the year 2011 
    plt.xticks(gas.Year[::3].tolist()+[2011])
    
    plt.legend(bbox_to_anchor=(1.02,1),loc='upper left',borderaxespad=0)
    plt.title('prices of gas overtime(USD)', fontdict={'fontsize':20, 'fontweight':'bold'})
    plt.savefig('real world example', dpi=300)
    
    
    
    
```

```python
plt.plot(gas.Year, gas.USA,'b.--', label='United state')
plt.plot(gas.Year, gas.Canada,'y.-', label='Canada')
plt.plot(gas.Year, gas.Australia, color='r',marker='*', label='Australia')
plt.plot(gas.Year, gas['South Korea'], marker='.', label='South Korea')

plt.xlabel('Year', fontdict={'fontsize':15})
plt.ylabel('US Dollar(s)',fontdict={'fontsize':15})

plt.xticks(gas.Year[::3])


plt.title('Gas prices overtime(USD)', fontdict={'fontsize':20, 'fontweight':'bold'})

plt.legend(bbox_to_anchor=(1.02,1), loc='upper left', borderaxespad=0)

plt.show()
plt.savefig('graph_2', dpi=300)
```

## Histogram

```python
bins = [30,40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins= bins, color='#21130d')
plt.xticks(bins)

plt.xlabel('Skill Level')
plt.ylabel('number of players')
plt.title('Distribution of players skills in FIFA 2018')
plt.show()
plt.savefig('histogram_1.png')
```

## pie Chart

```python
fifa.head(5)
Left= fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
Right= fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]

labels= ['Left', 'Right']
colors=['#063970','#1e81b0']
plt.pie([Left,Right], labels= labels, colors=colors, autopct='%.2f %%',)
plt.title('fifa foot preference', fontdict={'fontsize':20, 'fontweight': 'bold', 'fontname':'cursive'})
plt.show()

plt.savefig('pie chart.png')
```

## pie chart 2


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
fifa =pd.read_csv('fifa_data.csv')
```

```python
fifa.Weight=[int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light= fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight <200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium, medium, medium_heavy, heavy]


labels =['under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (.5,.2,0,0,.5)
plt.title('Weight Distribution of FIFA Players(lbs)',fontdict={'fontsize':20, 'fontweight':'bold'})

plt.pie(weights, labels = labels, autopct = '%.2f %%', explode= explode, startangle = 0)
plt.legend(bbox_to_anchor= (1.5,1), loc = 'upper left', borderaxespad = 1, title = 'players weight')
plt.show()
             
             
```

```python
barcelona=fifa.loc[fifa.Club=='FC Barcelona']['Overall']
madrid=fifa.loc[fifa.Club== 'Real Madrid']['Overall']

labels = ['FC Barcelona', 'Real Mdrid']
boxes=plt.boxplot([barcelona, madrid], labels=labels, patch_artist= True,)


for box in boxes['boxes']:
    # set edge color
    box.set(color='#4286f4', linewidth=2,)
    
    # change fill color
    box.set(facecolor='#e0e0e0')
    
    plt.title('Professional Soccer Team Comparison', fontdict={'fontsize':16, 'fontweight':'bold'})
    plt.ylabel('FIFA Overall Rating')

plt.show()
```
