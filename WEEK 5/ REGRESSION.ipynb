# Regression vs Correlation
Regression measures the degree a variable affects another variavble. for example , we can say the more education you have the more your salary increases. your regression must make sense, it does not make sense to say the more money you have the more education you want to have.
While correlation measure the degree of relationship between two variables

# Simple Linear Regression

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
data= pd.read_csv('1.01.+Simple+linear+regression.csv')
```

```python
data
data.describe()
```

```python
y= data['SAT']
x1= data['GPA']
```

```python
plt.scatter(x1,y, marker= '*', color= 'black', label = 'student')

yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat, lw=4, color='red', label='regression line')

plt.title('Simple regression', fontdict={'fontsize':30, 'fontname':'Helvetica'})
plt.ylabel('SAT', fontsize=20)
plt.xlabel('GPA')
plt.legend()
plt.show()
```

```python
import statsmodels.api as sm

x= sm.add_constant(x1)
result= sm.OLS(y,x).fit()
result.summary()
```

# Dummy Variable
The dummy variable is use to include categorical data into a regression model 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import statsmodels.api as sm
import seaborn as sns 
sns.set()
```

```python
data = pd.read_csv('1.03.+Dummies.csv')
```

```python
data
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```
