# WEEK6:DAY5- Introduction to Optimization
### cost function:
cost function is the measurement of how wrong the model is in terms of estimating the relationship between 
indepenedent and dependent variable (x and y).
### Gradient Descent:
 This is an efficient optimization algorithm that attemps to find a local or global minima of a function.
 Gradient descent enables the model to learn from the gradient or direction that the model should take in order to reduce error (ie the difference btw the predicted values and the actual values)

```python
# import the libraries
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.metrics import mean_squared_error as mse
```
